## ***Multiple Linear Regression.***



***`Multiple Linear Regression`*** is very similar to Simple Linear Regression, but this method is used to explain the relationship between: 

* One continuous response (dependent) variable.

* Two or more predictor (independent) variables.


$ \ $

$$
\begin{cases}
Y_1: Response \ Variable,\\ \\
X_{1} :Predictor\ Variable \ 1,\\ \\
X_{2} : Predictor\ Variable \ 2,\\ \\
X_{3} : Predictor\ Variable \ 3,\\ \\
X_{4} : Predictor\ Variable \ 4,\\ \\
. \\
. \\
. \\ \\
X_{n} : Predictor\ Variable \ n.
\end{cases} 
$$

$ \ $

$$
\begin{cases}
a: intercept,\\ \\
b_{1} :coefficients \ of\ Variable \ X_{1},\\ \\
b_{2} :coefficients \ of\ Variable \ X_{2},\\ \\
b_{3} :coefficients \ of\ Variable \ X_{3},\\ \\
b_{4} :coefficients \ of\ Variable \ X_{4},\\ \\
. \\
. \\
. \\ \\
b_{n} :coefficients \ of\ Variable \ X_{n}.
\end{cases}
$$

$ \ $

The equation is given by:

$$Y_{1} = a + b_{1} X_{1} + b_{2} X_{2} + b_{3} X_{3} + b_{4} X_{4}+\cdots+ b_{n} X_{n}. $$


$ \ $

# ***Residual Plot.***

A good way to visualize the variance of the data is to use a ***`residual plot`*** , but:


$ \ $

$$\text{What is a residual?}$$

$ \ $

The difference between the observed value $Y$ and the predicted value $Y_{1}$ is called the residual ($e=Y-Y_1$ / error). When we look at a regression plot, the residual is the distance from the data point to the fitted regression line.

---

$ \ $

$$\text{So what is a residual plot?}$$

$ \ $

A residual plot is a graph that shows the residuals on the vertical y-axis and the independent variable on the horizontal x-axis.

----

$ \ $

$$\text{What do we pay attention to when looking at a residual plot?}$$

$ \ $

* We look at the spread of the residuals. If the points in a residual plot are randomly spread out around the x-axis, then a linear model is appropriate for the data.

* Why is that? Randomly spread out residuals means that the variance is constant, and thus the linear model is a good fit for this data.


$ \ $

----

## ***Example (residual plot).***

Construct the residual plot between ***`'highway-mpg'`*** and ***`'price'`***.

````python
#(1) ancho de la imagen.
width = 12

#(2) largo de la imagen.
height = 10

#(3) configuramos el tamaÃ±o de la imagen teniendo en cuenta el ancho y largo descritos previamente.
plt.figure(figsize=(width, height))

#(4) We built the residual plot with the dataframe df using the columns 'highway-mpg" and  "price".
sns.residplot(x=df['highway-mpg'], y=df['price'])
plt.show()
````

$ \ $

$$ \text{What is this plot telling us?}$$

$ \ $

We can see from this residual plot that the residuals are not randomly spread around the x-axis, leading us to believe that maybe a non-linear model is more appropriate for this data.

# ***Polynomial Regression and Pipelines (tuberÃ­as).***


***`Polynomial regression`*** is a general case of the general linear regression model or multiple linear regression models. We get non-linear relationships by squaring or setting higher-order terms of the predictor variables. There are different orders of polynomial regression:

* ***`Quadratic`*** - 2nd Order.
$$Y_1 = a + b_1 X +b_2 X^2.$$

* ***`Cubic`*** - 3rd Order</b></center>
$$Y_1 = a + b_1 X +b_2 X^2 +b_3 X^3.$$

* ***`Higher-Order:`***
$$
Y_1 = a + b_1 X +b_2 X^2 +b_3 X^3+\cdots + b_n X^n.$$


$ \ $

````python
# We construct a function called plot_pol (plot polinomy / grafica polinomica).
def plot_pol(model, independent_variable, dependent_variabble, Name):

    #(1) realizamos una particiÃ³n con 100 elementos del intervalo [15,55].
    x_new = np.linspace(15, 55, 100)

    #(2) ejecutamos el modelo seleccionado sobre el arraw x_new.
    y_new = model(x_new)

    #(3) Realizamos un plot usando la variable dependiente Y y la variable independiente x, los puntos muestrales y el modelado.
    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')

    #(4) Describimos el tÃ­tulo.
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')

    #(5) GrÃ¡fica
    ax = plt.gca()

    #(6) colores grÃ¡fica
    ax.set_facecolor((0.898, 0.898, 0.898))

    #(7)  GrÃ¡fica
    fig = plt.gcf()

    #(8) Describimos las etiquetas de los ejes
    plt.xlabel(Name)    
    plt.ylabel('Price of Cars')

    #(9) Mostramos el resultado en pantalla y cerramos.
    plt.show()
    plt.close()
````

# ***Remark (description of polynomial functions of several variables).***

The analytical expression for Multivariate Polynomial function gets complicated. 

* Polynomial function of two - variables:  $ \ $ $Y_1=\sum \limits_{i,j} a_{(i,j)}X^{i}X^{j}$.

* Polynomial function of three -  variables:  $ \ $ $Y_1=\sum \limits_{i,j,k} a_{(i,j,k)}X^{i}X^{j}X^{k}$.

* Polynomial function of $ n$ - variables:  $ \ $ $Y_1=\sum \limits_{i_{1},\dots,i_{n}} a_{(i_{1},\dots,i_{n})}X^{i_{1}}\cdots X^{i_{n}}$.

# ***Pipelines (canalizaciones).***

***`Data Pipelines`*** (canalizaciones de datos) simplify the steps of processing the data. We use the module Pipeline (modelo canalizado) to create a pipeline (canal). We also use `StandardScaler` as a step in our pipeline (como un paso en nuestra canalizaciÃ³n).




````python
#(1) importamos pipeline (canal) de sklearn.pipeline.
from sklearn.pipeline import Pipeline

#(2) importamos StandardScaler (escaladora estÃ¡ndar) de sklearn.preprocessing.
from sklearn.preprocessing import StandardScaler
````

$ \ $

$(1)$ We create  a list of tuples including the name of the model or estimator and its corresponding constructor.


````python
#(1) name of the "model" = LinearRegression().
#(2) PolynomialFeatures(include_bias=False) is the constructor (the creator of features).
#(3) we describe the "scale" using "StandardScaler".
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model',LinearRegression())]
````

$ \ $

$(2)$ We input (describimos como entrada) the list as an argument to the pipeline  constructor (constructor del canal):

# ***Measures for In-Sample Evaluation.***

When evaluating our models, not only do we want to visualize the results, but we also want a quantitative measure to determine how accurate (preciso) the model is. Two very important measures that are often used in Statistics to determine the accuracy  of a model are:

* R^2 / R-squared.

* Mean Squared Error (MSE).

$ \ $

## ***`R-squared.`***

***`R - squared`***, also known as the coefficient of determination, is a measure to indicate how close the data is to the fitted regression line. The value of the ***R-squared*** is the percentage of variation of the response (response) variable $y$ that is explained by a linear model.

$ \ $

## ***`Mean Squared Error (MSE / Mean Squared Error).`***

The ***`Mean Squared Error (MSE)`*** measures the average of the squares of errors. That is, the difference between actual value $y$ and the estimated value $y_{1}$.


$ \ $ 

----

## ***Model 1: Simple Linear Regression.***

### ***`Example (R^2 of two sets of data)`***

Let's calculate the coefficient of determination ($R^2$) of $X$ and $Y$.

````python
#(1) We remember that X and Y have the following descrption.
X = df[["normalized-losses","highway-mpg"]]
Y = df[["price"]]

#(2) Remember that lm is given as:
lm=LinearRegression()

#(3) Encajamos X y Y.
lm.fit(X, Y)

#(3) Find the R^2
resultado = lm.score(X, Y)
print('The R-square is: ', resultado )
````

$ \ $

We can say that $49.659 \%$ of the variation of the price is explained by this simple linear model "horsepower_fit".


$ \ $

### ***`Example (mean squared error of a set of data)`***

Let's calculate the MSE of $X$ with the following steps:

$ \ $

$(1)$ We can predict the output $Y_1$ using the predict method, where $X$ is the input variable:


````python
#(1) We remember the definition of X and Y
X = df[["normalized-losses","highway-mpg"]]
Y = df[["price"]]

#(2) We creat a linear regression object called lm
lm = LinearRegression()

#(3) We remember that previously we fit X and Y.
lm.fit(X, Y)

#(4) We predict X to get some aprroximation of Y
Y_1=lm.predict(X)

#(5) We show on the screen our result
print('The output of the first four predicted value is: ', Y_1[0:4])
````

$ \ $

$(2)$ Let's import the function ***`mean_squared_error`*** from the module metrics:


````python
from sklearn.metrics import mean_squared_error
````

$ \ $

$(3)$ We can compare the predicted results with the actual results:


````python
#(1) We construct the mean squared error between Y_1 and Y=df['price'].
mse = mean_squared_error(Y, Y_1)

#(2) We show on the screen our previous result.
print('The mean square error of price and predicted value is: ', mse)
````
